#### Data Cleaning (Replace '--' to NaN)

```python
data.head(20)
```

```python
data=data.replace('--',np.nan, regex=True)
```

```python
data.head(20)
```

#### Check Null Values In The Dataset

```python
data.isnull().sum()
```

```python
per_missing=data.isnull().sum()*100/len(data)
```

```python
per_missing
```

```python
sns.heatmap(data.isnull())
```

```python
data.dropna(axis=0,inplace=True)
```

```python
sns.heatmap(data.isnull())
```

```python
data.isnull().sum()
```

#### Data Cleaning [ Rank Column ]

```python
data.head(100)
```

```python
data.dtypes
```

```python
data['Rank']=data['Rank'].str[0:-2]
```

```python
data['Rank']=data['Rank'].str.replace(',','').astype('int')
```

```python
data.dtypes
```


#### Data Cleaning [ Video Uploads & Subscribers ]

```python
data.head(20)
```

```python
data.info()
```

```python
data['Video Uploads']=data['Video Uploads'].astype('int')
```

```python
data['Subscribers']=data['Subscribers'].astype('int')
```


#### Data Cleaning [ Grade Column ]

```python
data['Grade'].unique()
```

```python
data['Grade']=data['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
```

```python
data.head(20)
```

#### Find Average Views For Each Channel

```python
data['AvgViews']=data['Video views']/data['Video Uploads']
```

```python
data.head()
```

#### Find Out Top Five Channels With Maximum Number of Video Uploads

```python
data.sort_values(by='Video Uploads',ascending=False).head()
```

#### Find Correlation Matrix

```python
df=pd.DataFrame(data)
```

```python
numberValues=df.select_dtypes(include=['number'])
```

```python
correlationMatrix=numberValues.corr()
correlationMatrix
```

```python
import matplotlib.pyplot as plt
```

```python
sns.heatmap(correlationMatrix,annot=True,cmap='coolwarm')
plt.show()
```
